```python
#Nominal / OHE >>binary vector for each categories
# example : single , married ,saperated
# single [1,0,0]
# married [0,1,0]
# saperated[0,0,1]

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df=pd.DataFrame({'status':['single','married','saperated','single','single','married','married']})
df
```

```python
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df[['status']]).toarray()
encoder.get_feature_names_out()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

now we can remove any one column from an the result will be still same eg if we remove status_single column then in row 0 the status_married = 0 and status_saperated is also 0 its simply indicates that status_single is 1 so in conclusion its n-1 i.e we can remove any column and the result will still be the same

```python
lst=['single','married','saperated','single','single','married','married']
dummies = pd.get_dummies(lst,dtype=int)
dummies
```

```python
pd.concat([pd.DataFrame(lst),dummies],axis=1)
```

```python
#label encoding>>Assigns unique label to categories
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
label_encoder.fit_transform(df['status'])
```

```python
# if new data comes
label_encoder.transform([['single']])
```

```python
#ordinal encoding
#High school : 1
#Graduate : 2
#pd:3
#Phd:4
from sklearn.preprocessing import OrdinalEncoder
df=pd.DataFrame({'qualification':['HS','GR','PG','PHD','HS','HS','PHD','PG']})
df
```

```python
encoder=OrdinalEncoder(categories=[['HS','GR','PG','PHD']])
encoder.fit_transform(df[['qualification']])
```

```python
#new data
encoder.transform([['PG']])
```

```python
#Target guided ordinal encoding
#relationship with target variables
#A lot of unique categories

#Replace the category with the mean and median of respective group target

df = pd.DataFrame({'time':['lunch','breakfast','dinner','lunch','breakfast','dinner','lunch','breakfast','dinner'],'total_bill':[120,130,90,125,150,190,160,180,189]})
df
```

```python
mean_price =df.groupby('time')['total_bill'].mean().to_dict()
mean_price
```

```python
df['time_encoded']=df['time'].map(mean_price)
df
```
