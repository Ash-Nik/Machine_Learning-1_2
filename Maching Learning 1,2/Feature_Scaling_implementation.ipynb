```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#festure scaling
#standardisation >> mu=0 and sigma =1
#Normalization (minmax scaling) >> data comes between 0 and 1
# unit vector
```

```python
df=sns.load_dataset('tips')
df
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
std_data=[]
for i in list(df['total_bill']):
  zscore=(i-mean)/std
  std_data.append(zscore)

std_data
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()
scalar.fit(df[['total_bill']]) # calculate mu and sigma

scalar.transform(df[['total_bill']])
```

```python
pd.DataFrame(scalar.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
#Normailization / minmax scaling
from sklearn.preprocessing import MinMaxScaler

minmax=MinMaxScaler()
pd.DataFrame(minmax.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
# unit vector >>
from sklearn.preprocessing import normalize
uv=normalize(df[['total_bill']])
pd.DataFrame(uv)
```
